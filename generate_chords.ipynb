```python
# 导入包
from colorChord import *
from utils import *
```

```python
color_chord_ls = generate_chords_by_color(['G', 'E', 'C'],
                                            [1, 13, 9, 8, 6, 5],
                                            normalize=False,
                                            max_chord_level=4,
                                            color_diff_range=(0, 20),
                                            random_seed=42)
```

```python
# 开始分析
color_change_ls = [0]
tension_change_ls = [0]
freshness_ls = [0]

for i in range(len(color_chord_ls) - 1):
    a = color_chord_ls[i]
    b = color_chord_ls[i+1]
    color_change_ls.append(Chord.get_color_change色彩变化(a, b))
    tension_change_ls.append(Chord.get_tension_change紧张度变化(a, b))
    freshness_ls.append(Chord.get_fressness新鲜度(a, b))
```

```python
# 生成dataframe
import pandas as pd
df = pd.DataFrame({
    'color_change': color_change_ls,
    'tension_change': tension_change_ls,
    'freshness': freshness_ls})
```
